Columns

1.  **Airline** - This contains the name of the airline company.
2.   **Date_of_Journey**- This contain the journey starting date
3.   **Source**-  This contain the place frome where the journey started.
4.   **Destination** - This contain places of destination
5.   **Route**-  This contain the route of flight
6.   **Dep_Time**- This contaion the departure time of flight.
7.   **Arrival_Time**- This contain the arrival time of flight.
8.   **Duration**- This is the time taken by plane to reach destination.
9.   **Total_Stops**- This contains the number of stops before reaching the destination.
10.  **Addtional_Info**- This contain aditional information of flight.
11.  Price- This contain the price of the flight

working on 'Arrival_Time' column

```python
# Convert Arrival_time into datetime and extract values

# Extracting Hours
train_data['Arrival_hour']=pd.to_datetime(train_data.Arrival_Time).dt.hour

# Extracting Minutes
train_data['Arrival_min']=pd.to_datetime(train_data.Arrival_Time).dt.minute
```

```python
# Removing Arrival_Time column
train_data.drop('Arrival_Time',axis=1,inplace=True)
```

```python
# value distribution of Arrival_min can be visualize
sns.distplot(train_data['Arrival_min'])
sns.boxplot(y=train_data['Arrival_min'])
plt.show()
```

```python
# to make hour and minute columns separately from Duration column

duration = list(train_data['Duration'])

for i in range(len(duration)):
  if len(duration[i].split()) !=2:                # checks if duration contain only hour or min
    if 'h' in duration[i]:
      duration[i] = duration[i].strip() + ' 0m'   # add 0 minute
    else:
      duration[i] = '0h ' + duration[i]           # add 0 hour
  else:
    continue

duration_hour=[]
duration_min=[]
for i in range(len(duration)):
  duration_hour.append(int(duration[i].split('h')[0]))                     # Extract hour from duration
  duration_min.append(int(duration[i].split()[1].split('m')[0]))           # Extract minute from duration
```

```python
# Make columns and assigning the values

train_data['Duration_hours']=duration_hour
train_data['Duration_mins']=duration_min
```

```python
train_data.drop(columns=['Duration'],inplace=True)
```

```python
train_data.head()
```

```python
# Value distribution
train_data.Duration_hours.value_counts()
```

```python
# value distribution of Duration_hours can be visualize
sns.distplot(train_data['Duration_hours'])
sns.boxplot(y=train_data['Duration_hours'])
plt.show()
```

```python
# value distribution of Duration_mins can be visualize
sns.distplot(train_data['Duration_mins'])
sns.boxplot(y=train_data['Duration_mins'])
plt.show()
```
